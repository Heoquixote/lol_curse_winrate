한계점: 데이터의 신뢰도가 부족. 게임하는 유저 말고, 게임하면서 욕설을 한 판과 그렇지 않고 하는 판을 비교해봐야 하지 않나 싶었음. 욕설하는 유저든, 욕설 안하는 유저든, 본인의 티어에 있다면 결국 승률은 50% 수렴하는것 아닌가 싶음. 욕설을 '했던 유저' 기준으로 볼게 아니라 욕설을 '한 게임'을 기준으로 보는게 좋지 않았을까 생각함.

아쉬운 점: 시각화가 아직 어려움.

좋았던 점: 직접 크롤링 해서 데이터셋을 만드는 경험을 해봤다는 점. Numpy와 Pandas를 적극 활용하여 데이터 프레임을 필요에 따라 조작해 보았음.

공유할 만한 insight: 욕설하는 유저의 데이터라고 모두 승률이 50% 이하가 되는 것이 아니었다. 분석을 할 때에 여러 가정이 있었음.
욕설을 하는 유저일수록 중요한 게임인 랭크 게임에서 자신이 포기하거나, 팀원을 포기하게 한다고 생각하여 졌을 경우 게임 시간이 짧을 거라고 예상하였지만, 막상 뜯어보면 그렇지 않았음.
오히려 티어 구간에서 승률 차이만 보였을 뿐이었다. 각 구간에서 티어 1인 유저와 티어 4인 유저의 승률 차이가 약 3% 났었다.

뿐만 아니라, 비록 게임 시간은 비슷하였지만, 욕설하는 유저의 솔랭 승률은 48%였지만, 비교적 자유로운 분위기인 자랭 승률은 52%로 4%의 차이가 존재했음. - > 이를 욕설하는 유저는 솔랭에서 욕설을 했기에 승률이 낮고, 비교적 부담이 덜한 자랭에서는 욕설을 안했기에 승률이 52%라는 해석을 하기에는 무리가 있음.
이 외에 다른 조건들로 분석해 봤을때 크게 의미있는 차이가 보이지 않았음.

솔랭과 자랭에서의 승률 차이의 원인은 무엇인지 모르겠으나, 티어1에서 욕설하는 유저는 진심으로 승급을 원하기에 게임을 진심으로 대하다 보니 욕설을 하게 되고 승률도 약 51%가 되고, 티어 4인 유저들은 이미 그 승급이라는 목적을 달성했기에 코앞의 목표가 없어 승률이 48%로 나오는거로 예측된다.
