```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
```

```python
df.columns = ['id', 'tier', 'mode', 'W/L', 'champion', 'position', 'Kill', 'Death', 'Assist', 'kill participance', 'KDA']
df
```

```python
# 전처리
# 1. 게임 전적 중, 무조건 승일수 밖에 없는 '봇전', '커스텀', 그리고 어떤 게임인지 모르는 '기타' 모드는 삭제

df = df[df['mode'] != '봇전']
df
```

```python
df = df[df['mode'] != '커스텀']
df = df[df['mode'] != '기타']
df
```

```python
df['mode'].value_counts()
```

```python
# W/L에서 리겜 된 'RE'는 제거

df = df[df['W/L'] != 'RE']
df
```

```python
df['W/L'].value_counts()
```

```python
# 킬 관여율 column인 'kill participance' 칼럼에서 킬관여율 한자리 숫자일때 X% 로 표현 된 것 모두 숫자로 치환하고
# 데이터 타입을 object에서 int로 바꾸기

df = df.replace('(.*)%', r'\1', regex=True)
df['kill participance'] = pd.to_numeric(df['kill participance'])
df['kill participance'].dtype
```

```python
df
```

```python
df.info()
```

```python
# 욕설 유저들의 티어 분포

sns.set(rc = {'figure.figsize': (18, 5)})
sns.countplot(data = df, x = 'tier', order = ['Unranked', 'Iron 4', 'Iron 3', 'Iron 2', 'Iron 1', 'Bronze 4', 'Bronze 3', 'Bronze 2', 'Bronze 1', 'Silver 4', 'Silver 3', 'Silver 2', 'Silver 1', 'Gold 4', 'Gold 3', 'Gold 2', 'Gold 1', 'Platinum 4', 'Platinum 3', 'Platinum 2', 'Platinum 1', 'Diamond 4', 'Diamond 3', 'Diamond 2', 'Diamond 1', 'Master'])
```

```python
# op.gg 피셜 랭크게임 유저중 브실골 유저 % : 84.24%

len(set(df[(df['tier'] == 'Bronze 4') |(df['tier'] == 'Bronze 3') |(df['tier'] == 'Bronze 2') |(df['tier'] == 'Bronze 1') |(df['tier'] == 'Silver 4')|(df['tier'] == 'Silver 3')|(df['tier'] == 'Silver 2')|(df['tier'] == 'Silver 1')|(df['tier'] == 'Gold 4')|(df['tier'] == 'Gold 3')|(df['tier'] == 'Gold 2')|(df['tier'] == 'Gold 1')]['id']))/ len(set(df[df['tier'] != 'Unranked']['id']))

# 욕설 박제된 닉네임들의 브실골 비율은 랭크 있는 게임 기준 86.41%로 미세하게 많다.
```
