```python
# 아이언
perc_i = len(df[(df['tier']=='Iron 4')|(df['tier']=='Iron 3')|(df['tier']=='Iron 2')|(df['tier']=='Iron 1')]['id'].unique()) / len(df[df['tier'] != 'Unranked']['id'].unique())
print(perc_i)
print('데이터 속 아이언의 비율은 {:.2f}%이고, 이는 전체의 그것의 {:.2f}% 수치이다.'.format(perc_i*100, (perc_i*100)/4.04 * 100))
```

```python
# 브론즈
perc_b = len(df[(df['tier']=='Bronze 4')|(df['tier']=='Bronze 3')|(df['tier']=='Bronze 2')|(df['tier']=='Bronze 1')]['id'].unique()) / len(df[df['tier'] != 'Unranked']['id'].unique())
print(perc_b)
print('데이터 속 브론즈의 비율은 {:.2f}%이고, 이는 전체의 그것의 {:.2f}% 수치이다.'.format(perc_b*100, perc_b/23.47 * 10000))
```

```python
# 실버
perc_s = len(df[(df['tier']=='Silvere 4')|(df['tier']=='Silvere 3')|(df['tier']=='Silvere 2')|(df['tier']=='Silver 1')]['id'].unique()) / len(df[df['tier'] != 'Unranked']['id'].unique())
print(perc_s)
print('데이터 속 실버의 비율은 {:.2f}%이고, 이는 전체의 그것의 {:.2f}% 수치이다.'.format(perc_s*100, perc_s/34.24 * 10000))
```

```python
# 골드

perc_g = len(df[(df['tier']=='Gold 4')|(df['tier']=='Gold 3')|(df['tier']=='Gold 2')|(df['tier']=='Gold1')]['id'].unique()) / len(df[df['tier'] != 'Unranked']['id'].unique())
print(perc_g)
print('데이터 속 골드의 비율은 {:.2f}%이고, 이는 전체의 그것의 {:.2f}% 수치이다.'.format(perc_g*100, perc_g/26.55 * 10000))
```

```python
# 플레티넘
perc_p = len(df[(df['tier']== 'Platinume 4')|(df['tier']== 'Platinume 3')|(df['tier']== 'Platinume 2')|(df['tier']== 'Platinum 1')]['id'].unique()) / len(df[df['tier'] != 'Unranked']['id'].unique())
print(perc_p)
print('데이터 속 플레티넘의 비율은 {:.2f}%이고, 이는 전체의 그것의 {:.2f}% 수치이다.'.format(perc_p*100, perc_p/10.13 * 10000))
```

```python
# 다이아몬드
perc_d = len(df[(df['tier']=='Diamond 4')|(df['tier']=='Diamond 3')|(df['tier']=='Diamond 2')|(df['tier']=='Diamond 1')]['id'].unique()) / len(df[df['tier'] != 'Unranked']['id'].unique())
print(perc_d)
print('데이터 속 다이아의 비율은 {:.2f}%이고, 이는 전체의 그것의 {:.2f}% 수치이다.'.format(perc_d*100, perc_d/1.46 * 10000))
```

```python
# 마스터
perc_m = len(df[(df['tier']=='Master')]['id'].unique()) / len(df[df['tier'] != 'Unranked']['id'].unique())
print(perc_m)
print('데이터 속 마스터의 비율은 {:.2f}%이고, 이는 전체의 그것보다 {:.2f}% 수치이다.'.format(perc_m*100, perc_m/10.13 * 10000))
```
